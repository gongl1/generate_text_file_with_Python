```python
# 3 file size option to choose from below for testing purpose

# walden_clean_location = "walden_clean - Copy.txt"
# walden_clean_location = "walden_clean - short.txt"
walden_clean_location = "walden_clean.txt"

# Turn walden_clean.txt into walden_clean_noperiod.txt which include the word PERIOD rather than .

# output_file_location = "walden_clean_noperiod - Copy.txt"
# output_file_location = "walden_clean_noperiod - short.txt"
output_file_location = "walden_clean_noperiod.txt"

f = open(walden_clean_location, 'r', encoding='utf8')
lista = f.readlines()

# transform each line in original text file to new text file called "walden_clean_noperiod.txt"
with open(output_file_location, 'w', encoding='utf8') as output_file:
    for line in lista:
        line = line.replace(".", " PERIOD")
        # print(line)
        output_file.write(line)
```
